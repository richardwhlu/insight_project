```python
powerset_col = []

for index, row in data.iterrows():
    if row["food"]:
        if row["service"]:
            if row["price"]:
                if row["ambiance"]:
                    powerset_col.append("fspa")
                else:
                    powerset_col.append("fsp_")
            else:
                if row["ambiance"]:
                    powerset_col.append("fs_a")
                else:
                    powerset_col.append("fs__")
        else:
            if row["price"]:
                if row["ambiance"]:
                    powerset_col.append("f_pa")
                else:
                    powerset_col.append("f_p_")
            else:
                if row["ambiance"]:
                    powerset_col.append("f__a")
                else:
                    powerset_col.append("f___")
    else:
        if row["service"]:
            if row["price"]:
                if row["ambiance"]:
                    powerset_col.append("_spa")
                else:
                    powerset_col.append("_sp_")
            else:
                if row["ambiance"]:
                    powerset_col.append("_s_a")
                else:
                    powerset_col.append("_s__")
        else:
            if row["price"]:
                if row["ambiance"]:
                    powerset_col.append("__pa")
                else:
                    powerset_col.append("__p_")
            else:
                if row["ambiance"]:
                    powerset_col.append("___a")
                else:
                    powerset_col.append("____")
```

```python
data["powerset"] = powerset_col
```
