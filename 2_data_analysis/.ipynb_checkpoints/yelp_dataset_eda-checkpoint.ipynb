```python
import nltk
from nltk.stem.porter import *
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string
# stopwords.words('english')
```

```python
def process_reviews(text):
    stemmer = PorterStemmer()
    tmp = [stemmer.stem(x.lower()) 
           for x in word_tokenize(text)
           if x.lower() not in stopwords.words("english")
           and x.lower() not in string.punctuation]
    return " ".join(tmp)
```

```python
test = sample_reviews["text"].apply(process_reviews)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tf_vectorizer = CountVectorizer(max_df=0.95,
                                min_df=2,
                                max_features=1000)
tf = tf_vectorizer.fit_transform(test)
```

```python
tf_feature_names = tf_vectorizer.get_feature_names()
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.95,
                                   min_df=2,
                                   max_features=1000)
tfidf = tfidf_vectorizer.fit_transform(test)
```

```python
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
```

```python
from sklearn.decomposition import NMF, LatentDirichletAllocation as LDA
```

```python
num_topics = 20
```

```python
nmf = NMF(n_components=num_topics,
          random_state=10191994,
          alpha=.1,
          l1_ratio=.5,
          init="nndsvd").fit(tfidf)
```

```python
lda = LDA(n_components=num_topics,
          max_iter=5,
          learning_method="online",
          learning_offset=50,
          random_state=10191994).fit(tf)
```

```python
def display_topics(model, feature_names, num_top_words):
    for topic_index, topic in enumerate(model.components_):
        print("Topic {}:".format(topic_index))
        print(" ".join([feature_names[i]
                       for i in topic.argsort()[:-num_top_words - 1:-1]]))
```

```python
display_topics(nmf, tfidf_feature_names, 10)
```

```python
display_topics(lda, tf_feature_names, 10)
```

```python
doc_topic = lda.transform(tf)
```

### Take a look at most common stems

```python
def get_most_common_words_across_sample_reviews(random_seed, num_words):
    sample_reviews = select_reviews(random_seed, 2000)
    tmp_df = sample_reviews["text"].apply(process_reviews)
    tf_vectorizer = CountVectorizer(max_df=0.95,
                                    min_df=2,
                                    max_features=1000)
    tf = tf_vectorizer.fit(tmp_df)
    tf2 = tf.transform(tmp_df)
    term_sums = tf2.sum(axis=0)
    words_freq = [(word, term_sums[0, index])
                  for word, index in tf.vocabulary_.items()]
    return sorted(words_freq, key=lambda x: x[1], reverse=True)[:num_words]
```

```python
get_most_common_words_across_sample_reviews(10191994, 30)
```

```python
from collections import defaultdict
```

```python
categories_dict = defaultdict(int)
```

```python
for index, row in business.iterrows():
    categories = row["categories"]
    try:
        list_of_categories = categories.split(", ")
        for category in list_of_categories:
            categories_dict[category] += 1
    except:
        pass
```
