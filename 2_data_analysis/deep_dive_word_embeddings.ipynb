```python
import copy
import string
from nltk.corpus import stopwords
from nltk.stem.porter import *
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity

def preprocess_reviews(review_text, stem, join=0):
    """Clean up the raw review text a little.

    Args:
        review_text - raw review text from csv
        stem - stem tokens or not
        join - join the stems into sentence or not

    Returns:
        cleaned_sentence - tokenized and stemmed sentence
    """
    stemmer = PorterStemmer()
    if stem:
        cleaned_sentence = [stemmer.stem(x.lower()) 
               for x in word_tokenize(review_text)
               if x.lower() not in stopwords.words("english")
               and x.lower() not in string.punctuation]
        if join:
            cleaned_sentence = " ".join(cleaned_sentence)
    else:
        cleaned_sentence = [x.lower() 
               for x in word_tokenize(review_text)
               if x.lower() not in stopwords.words("english")
               and x.lower() not in string.punctuation]
    return cleaned_sentence


def test_embedding_features(text, word, wordlist=[]):
    clean_text = preprocess_reviews(text, 0)
    clean_vec = copy.deepcopy(model.wv["word"])
    for word_parse in wordlist:
        clean_vec -= copy.deepcopy(model.wv[word_parse])
    return sorted([(x, float(cosine_similarity(
                                         clean_vec.reshape(1, -1),
                                         model.wv[x].reshape(1, -1)))) for x in clean_text],
                  key=lambda x: -x[1])

def test_embedding_features2(text, wordlist):
    clean_text = preprocess_reviews(text, 0)
    most_similar_list = []
    for word in clean_text:
        tmp = []
        for word2 in wordlist:
            tmp.append((word, word2, model.wv.similarity(word, word2)))
        
        tmp2 = sorted(tmp, key=lambda x: -x[-1])
        print(tmp2)
        most_similar_list.append(tmp2[0])
        
    return most_similar_list
```
