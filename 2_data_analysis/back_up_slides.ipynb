```python
from sklearn.metrics import precision_recall_fscore_support, roc_curve, confusion_matrix, roc_auc_score
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4)
```

```python
def model_results(model, test_x, test_y,):
    pp.pprint(precision_recall_fscore_support(test_y, model.predict(test_x)))
    pp.pprint(roc_auc_score(test_y, model.predict_proba(test_x)[:, 1]))
    pp.pprint(confusion_matrix(test_y, model.predict(test_x)))
```
